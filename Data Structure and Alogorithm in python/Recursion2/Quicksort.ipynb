```python
def partition(a,si,ei):
    count = 0
    i=si
    j=ei
    pivot = a[si]
    k=si+1
    while(k<=ei):
        if(pivot>a[k]):
            count=count+1
        k = k + 1
    a[si+count],a[si]=a[si],a[si+count]
    while i<j:
        if(a[i]<a[count]):
            i=i+1
        elif(a[j]>a[count]):
#             print(j)
            j = j -1
        else:
            a[i],a[j]=a[j],a[i]
            i = i + 1
            j = j - 1
    return count        
    
```

```python
def quicksort(a,si,ei):
    if(si>=ei):
        return
    pi = partition(a,si,ei)
    print(a,pi)
    quicksort(a,si,pi-1)
    quicksort(a,pi+1,ei)
```
